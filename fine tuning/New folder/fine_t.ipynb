```python
import json, fitz, sys, os
from pathlib import Path
docs = []
folder = '.'
for pdf in Path(folder).rglob('*.pdf'):
    print(f' {pdf.name}')
    try:
        doc = fitz.open(pdf)
        text = ''.join(page.get_text() for page in doc)
        doc.close()
        if text.strip(): docs.append({'text': text.strip(), 'filename': str(pdf)})
    except: pass
with open('fine_tune_dataset.jsonl', 'w', encoding='utf-8') as f:
    for d in docs: f.write(json.dumps(d, ensure_ascii=False)+'\n')
print(f'\n {len(docs)} PDFs → fine_tune_dataset.jsonl')
```
