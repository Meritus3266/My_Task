```python
# using median to determine the missing value in age and footlength

age_med = df['age'].median()
print(age_med)

footlength_med = df['footlgth'].median()
print(footlength_med)


df['age'] = df['age'].fillna(age_med)
df['footlgth'] = df['age'].fillna(footlength_med)

# checking if there is any missing file after the correction
df.isna().sum()
```

```python
num = ['case', 'site', 'age', 'hdlngth', 'skullw', 'totlngth',
       'taill', 'footlgth', 'earconch', 'eye', 'chest', 'belly']
```

```python
df_num1 = pd.DataFrame({
    'mean': df[num].mean().round(2),
    'median': df[num].median().round(2),
    'mode': df[num].mode().iloc[0].round(2),
    'std': df[num].std().round(2),
    'variance': df[num].var().round(2)
})
```

```python
df_num1
```

```python
df_num2 = pd.DataFrame({
    'max': df[num].max().round(2),
    'min': df[num].min().round(2),
    'range': (np.subtract(df[num].max(), df[num].min())).round(2),
    '25th_percentile': df[num].quantile(0.25).round(2),
    '50th_percentile': df[num].quantile(0.50).round(2),
    '75th_percentile': df[num].quantile(0.75).round(2),
    'IQR': (np.subtract(df[num].quantile(0.75),df[num].quantile(0.25))).round(2),
    'skewness': df[num].skew(),
    'kurtosis': df[num].kurtosis()
    })
```
