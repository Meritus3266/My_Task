```python
from PyPDF2 import PdfReader
```

```python
def extract_text_pypdf2(path):
    reader = PdfReader(path)
    pages = []
    for p in reader.pages:
        pages.append(p.extract_text() or "")
    return pages  

# Usage
pages = extract_text_pypdf2("Python For Dummies.pdf")
print(len(pages), "pages loaded")
```

```python
def chunk_by_page(pages):
    chunks = []
    for i, page in enumerate(pages, start=1):
        chunks.append({
            "page_number": i,
            "text": page
        })
    return chunks
```
