```python


# Define matrices A and B
A = np.array([[2, -1],
              [0, 3]])

B = np.array([[1, 4],
              [-2, 5]])

# Define scalars
c = 2
d = 3

# Scalar multiplication
cA = c * A
print("cA =\n", cA)

dB = d * B
print("dB =\n", dB)

# (c + d) * A
cd_A = (c + d) * A
print("(c + d)A =\n", cd_A)

# Associativity of scalar multiplication: (c * d)A == c * (dA)
cdA = (c * d) * A
print("(c * d)A =\n", cdA)

dA = d * A
cdA_check = c * dA
print("c(dA) =\n", cdA_check)

# Distributivity over matrix addition: c(A + B) == cA + cB
A_plus_B = A + B
c_A_plus_B = c * A_plus_B
print("c(A + B) =\n", c_A_plus_B)

cB = c * B
cA_plus_cB = cA + cB
print("cA + cB =\n", cA_plus_cB)

# Distributivity over scalar addition: (c + d)A == cA + dA
cdA_again = cA + dA
print("cA + dA =\n", cdA_again)

# Transposes
A_T = A.T
print("Transpose of A:\n", A_T)

B_T = B.T
print("Transpose of B:\n", B_T)
```
