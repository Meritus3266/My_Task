### CLASS TASK TWO MATRIX

### Determining (AB)

```python
import numpy as np

A = np.array([[2, -1, 3],
              [0, 4, 1]])

print('A array shape:', A.shape)

B = np.array([[1, 2], [-2, 0],
              [5, -3]])

print('B array shape:', B.shape)

print(A @ B)
```

### Determining (BA)

```python
B = np.array([[1, 2], [-2, 0],
              [5, -3]])

print('B array shape:', B.shape)

A = np.array([[2, -1, 3],
              [0, 4, 1]])

print('A array shape:', A.shape)

print(B @ A)

```
